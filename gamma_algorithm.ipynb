```python
def symmetric_difference(G, H):
    
    # create new graph
    if not G.is_multigraph() == H.is_multigraph():
        raise nx.NetworkXError('G and H must both be graphs or multigraphs.')
    R = nx.create_empty_copy(G)
    R.name = "Symmetric difference of (%s and %s)"%(G.name, H.name)

    #if set(G) != set(H):
        #raise nx.NetworkXError("Node sets of graphs not equal")

    gnodes = set(G) # set of nodes in G
    hnodes = set(H) # set of nodes in H
    nodes = gnodes.symmetric_difference(hnodes)
    R.add_nodes_from(nodes)

    if G.is_multigraph():
        edges = G.edges_iter(keys=True)
    else:
        edges = G.edges_iter()
    # we could copy the data here but then this function doesn't
    # match intersection and difference
    for e in edges:
        if not H.has_edge(*e):
            R.add_edge(*e)

    if H.is_multigraph():
        edges = H.edges_iter(keys=True)
    else:
        edges = H.edges_iter()
    for e in edges:
        if not G.has_edge(*e):
            R.add_edge(*e)
    return R
```

```python
def Get_Segments_2_type(G,G1):
    T=symmetric_difference(G, G1)
    T1=nx.Graph()
    T1=T
    ed=[]
    for i in range(len(list(T.edges()))):
        if T.edges()[i][0] in G1.nodes() and T.edges()[i][1] in G1.nodes():
            ed.append(T.edges()[i])
        #T1.remove_edge(T1.edges()[i][0],T1.edges()[i][1])
    #print ed
    for i in range(len(ed)):
        T1.remove_edge(*(ed[i]))
    return T1
```

```python
def get_T2(T1,G,G1):
    ed=[]
    for i in range(len(list(T1.edges()))):
        if T1.edges()[i][0] not in G1.nodes() and T1.edges()[i][1] in G1.nodes() or T1.edges()[i][0]  in G1.nodes() and T1.edges()[i][1] not in G1.nodes() :
            ed.append(T1.edges()[i])
        #T1.remove_edge(T1.edges()[i][0],T1.edges()[i][1])
    #print ed
#for i in range(len(ed)):
    #T2.remove_edge(*(ed[i]))
    T2=nx.Graph()
    T2=Get_Segments_2_type(G,G1)
    for i in T2.nodes():
        if i in G1.nodes():
            T2.remove_node(i)
    f=list(nx.connected_components(T2))
    return f
#число сегментов второго типа
```

```python
def Get_all_segments(G,G1):
    T=symmetric_difference(G, G1)
    #print T.edges()
    T1=Get_Segments_2_type(G,G1)
    #print T1.edges()
    f=get_T2(T1,G,G1)
    #print 'f',f
    Sub={}
    Sub=Get_subgraphs_for_2_segments(f,T1)
    #print 'sub',Sub
    
    S=get_Segment_first_type(G,G1,T)
    #print 's',S
    Segments=dict((Sub.items())+(S.items()))
    for i in Segments.keys():
        if len(list(Segments[i].edges()))==0:
            Segments.pop(i)
    
#for i in Segments.keys():
        #print Segments[i].edges()
    return Segments
```

```python
def Getting_grans_as_graphs(arr):
    slovar_grans={}
    k=1
    G=nx.Graph()
    for i in range(len(arr)):
        G.add_cycle(arr[i])
        slovar_grans[k]=G
        k=k+1
        G=nx.Graph()
    return slovar_grans
        
```
